```python
from tkinter import *

master = Tk()

def ekraniOrtala(nesne):
    pencere_genisligi = 500
    pencere_yuksekligi = 300

    ekran_genisligi = nesne.winfo_screenwidth()
    ekran_yuksekligi = nesne.winfo_screenheight()

    x_kord = (ekran_genisligi - pencere_genisligi) // 2
    y_kord = (ekran_yuksekligi - pencere_yuksekligi) // 2
    
    nesne.geometry(f"{pencere_genisligi}x{pencere_yuksekligi}+{x_kord}+{y_kord}")
    
    
master.title("Giriş Ekranı")

Canvas(master,bg = "#E7CEA6").place(relx=0.0,rely=0.0,relheight=1,relwidth=1)

master.minsize(500,300)
master.maxsize(500,300)

ekraniOrtala(master)

kullanici_adi = Label(master,text = "Kullanıcı Adı: ",font="Consolas 12 bold",bg="#E7CEA6")
kullanici_adi.grid(row=0,column=0,padx=10,pady=10)
kullanici_adi_txt = Entry(master,font="Consolas 12 bold",bg="#5A96E3",foreground="#E7CEA6")
kullanici_adi_txt.grid(row=0,column=1,padx=10,pady=10)


kullanici_sifre = Label(master,text="Şifre: ",font="Consolas 12 bold",bg="#E7CEA6")
kullanici_sifre.grid(row=1,column=0,padx=10,pady=10)
kullanici_sifre_txt = Entry(master,font="Consolas 12 bold",bg="#5A96E3",foreground="#E7CEA6",show="*")
kullanici_sifre_txt.grid(row=1,column=1,padx=10,pady=10)

def sifreGoster():
    if kullanici_sifre_txt["show"]=="*":
        kullanici_sifre_txt["show"]=""
    else:
        kullanici_sifre_txt["show"]="*"

def kullaniciKontrol():
    if kullanici_adi_txt.get()=='veysel' and kullanici_sifre_txt.get()=='veysel':
        Label(master,text="Giriş Başarılı",font="Consolas 11 bold",foreground="Green",background="#E7CEA6").grid(row=3,column=1,padx=10,pady=10)
    else:
        Label(master,text="Giriş Başarısız",font="Consolas 11 bold",foreground="Red",background="#E7CEA6").grid(row=3,column=1,padx=10,pady=10)

giris_buton = Button(master,text="Giriş Yap",font="Consolas 12 bold",bg = "#0A6EBD",foreground= "#E7CEA6",command=kullaniciKontrol,activebackground="#A1C2F1")
giris_buton.grid(row=2,column=1,padx=10,pady=10)

sifre_gizle = Checkbutton(master,text="Şifre Göster",command=sifreGoster,background="#E7CEA6",font="Consolas 11 bold",foreground="#5A96E3")
sifre_gizle.grid(row=1,column=2,padx=10,pady=10)


master.mainloop()
```
